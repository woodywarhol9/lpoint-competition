```python
one=pd.read_csv("/content/drive/MyDrive/LDA_j/month/one.csv")
two=pd.read_csv("/content/drive/MyDrive/LDA_j/month/two.csv")
three=pd.read_csv("/content/drive/MyDrive/LDA_j/month/three.csv")
four=pd.read_csv("/content/drive/MyDrive/LDA_j/month/four.csv")
five=pd.read_csv("/content/drive/MyDrive/LDA_j/month/five.csv")
six=pd.read_csv("/content/drive/MyDrive/LDA_j/month/six.csv")
seven=pd.read_csv("/content/drive/MyDrive/LDA_j/month/seven.csv")
eight=pd.read_csv("/content/drive/MyDrive/LDA_j/month/eight.csv")
nine=pd.read_csv("/content/drive/MyDrive/LDA_j/month/nine.csv")
ten=pd.read_csv("/content/drive/MyDrive/LDA_j/month/ten.csv")
eleven=pd.read_csv("/content/drive/MyDrive/LDA_j/month/eleven.csv")
twelve=pd.read_csv("/content/drive/MyDrive/LDA_j/month/twelve.csv")
```

```python
one=one.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
two=two.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
three=three.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
four=four.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
five=five.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
six=six.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
seven=seven.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
eight=eight.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
nine=nine.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
ten=ten.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
eleven=eleven.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
twelve=twelve.groupby(['대상연월','요일','성별','나이']).sum().reset_index()
```
