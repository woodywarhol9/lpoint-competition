```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


demo = pd.read_csv("/content/mnt/MyDrive/LDA/data/LPOINT_BIG_COMP_01_DEMO.csv")
pdde = pd.read_csv("/content/mnt/MyDrive/LDA/data/LPOINT_BIG_COMP_02_PDDE.csv")
cop = pd.read_csv("/content/mnt/MyDrive/LDA/data/LPOINT_BIG_COMP_03_COP_U.csv")
pd_clac = pd.read_csv("/content/mnt/MyDrive/LDA/data/edit_pd_clac.csv")
br = pd.read_csv("/content/mnt/MyDrive/LDA/data/LPOINT_BIG_COMP_05_BR.csv")
lpay = pd.read_csv("/content/mnt/MyDrive/LDA/data/LPOINT_BIG_COMP_06_LPAY.csv")
```

```python
# 사용처 구분
pdde['deal_type'] = 0
cop['deal_type'] = 1
lpay['deal_type'] = 2
```

```python
data = pd.concat([pdde, cop, lpay])
data = data.merge(demo, on = "cust", how = 'left')
# zon_hlv 중복 방지
data.rename(columns = {"zon_hlv" : "zon_hlv_user"}, inplace = True)
data = data.merge(br, on = 'br_c', how = 'left')
# br_c null
data.drop(columns = ["cop_c_y"], inplace = True)
data.rename(columns = {"cop_c_x" : "cop_c"}, inplace = True)
data
```
